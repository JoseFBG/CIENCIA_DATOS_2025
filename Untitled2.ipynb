```python
# ===========================
# 0) PREPARACIÓN EN COLAB
# ===========================
!pip -q install pandas numpy requests matplotlib seaborn tqdm ratelimit tenacity

import os, time, math, json, statistics
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from ratelimit import limits, sleep_and_retry
from tenacity import retry, stop_after_attempt, wait_exponential

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 160)
sns.set(context="notebook", style="whitegrid")

# ===========================
# 1) CONFIGURACIÓN
# ===========================
API_KEY   = "RGAPI-7d605454-a856-4b43-b783-515ec27b1b0f"
riot_name = "G2 Charli"
riot_tag  = "0000"

# Para LATAM/NA/BR: americas. Ajusta si tu cuenta es EU/ASIA.
ROUTING   = "americas"        # para account-v1 y match-v5
PLATFORM  = "LA2"             # útil si luego usas summoner-v4
BASE_ACC  = f"https://{ROUTING}.api.riotgames.com/riot/account/v1"
BASE_M5   = f"https://{ROUTING}.api.riotgames.com/lol/match/v5"
HEADERS   = {"X-Riot-Token": API_KEY}

# ===========================
# 2) HELPERS (Requests/Rate Limit/Retry)
# ===========================
RATE_CALLS = 18
RATE_SEC   = 1

@sleep_and_retry
@limits(calls=RATE_CALLS, period=RATE_SEC)
def _safe_get(url, params=None):
    r = requests.get(url, headers=HEADERS, params=params, timeout=15)
    if r.status_code == 429:
        retry_after = int(r.headers.get("Retry-After", "2"))
        time.sleep(retry_after + 1)
        raise Exception("Rate limited, retrying...")
    if r.status_code >= 500:
        raise Exception(f"Server error {r.status_code}, retrying...")
    r.raise_for_status()
    return r.json()

@retry(wait=wait_exponential(multiplier=1, min=2, max=20), stop=stop_after_attempt(5))
def safe_get(url, params=None):
    return _safe_get(url, params=params)

# ===========================
# 3) OBTENER PUUID DESDE RIOT ID
# ===========================
def get_puuid(riot_name, riot_tag):
    url = f"{BASE_ACC}/accounts/by-riot-id/{riot_name}/{riot_tag}"
    data = safe_get(url)
    return data["puuid"]

puuid = get_puuid(riot_name, riot_tag)
print("PUUID:", puuid)

# ===========================
# 4) OBTENER LISTA DE MATCH IDS POR TIPO
# ===========================
def get_match_ids(puuid, type_filter="ranked", count=50, start=0):
    url = f"{BASE_M5}/matches/by-puuid/{puuid}/ids"
    params = {"type": type_filter, "start": start, "count": count}
    return safe_get(url, params=params)

# ranked (incluye Solo/Duo 420 y Flex 440) y normales
match_ids_ranked = get_match_ids(puuid, type_filter="ranked", count=100)
match_ids_normal = get_match_ids(puuid, type_filter="normal", count=100)
print("Ranked IDs:", len(match_ids_ranked), "| Normal IDs:", len(match_ids_normal))

# ===========================
# 5) DESCARGAR DETALLES + TIMELINE
# ===========================
def fetch_match(match_id, with_timeline=True):
    m = safe_get(f"{BASE_M5}/matches/{match_id}")
    t = None
    if with_timeline:
        try:
            t = safe_get(f"{BASE_M5}/matches/{match_id}/timeline")
        except Exception:
            t = None
    return m, t

# ===========================
# 6) DDragon Items (NOMBRES + MÍTICOS)  ✅ CORREGIDO
# ===========================
def load_items_catalog():
    # versión más reciente
    versions = requests.get("https://ddragon.leagueoflegends.com/api/versions.json", timeout=15).json()
    latest = versions[0]
    items = requests.get(
        f"https://ddragon.leagueoflegends.com/cdn/{latest}/data/en_US/item.json",
        timeout=15
    ).json()["data"]

    id2name = {}
    mythic_set = set()  # usar el mismo nombre que devolvemos

    for item_id, meta in items.items():
        iid = int(item_id)
        id2name[iid] = meta.get("name", str(iid))
        desc = meta.get("description", "")
        tags = meta.get("tags", [])
        # Detecta míticos: condición más fiable es "Mythic Passive" en la descripción.
        if "Mythic Passive" in desc or "Mythic" in tags:
            mythic_set.add(iid)

    return id2name, mythic_set

id2name, mythic_set = load_items_catalog()

# ===========================
# 7) PARSERS DE ESTADÍSTICAS DEL JUGADOR
# ===========================
QUEUE_NAMES = {
    420: "Solo/Duo",
    440: "Flex",
    400: "Normal Draft",
    430: "Normal Blind",
    450: "ARAM",
    700: "Clash"
}

LANE_BOXES = {
    "TOP":   ((2000, 12000), (12000, 15000)),
    "MID":   ((5500, 9500),  (5500, 9500)),
    "BOT":   ((12000, 15000),(2000, 12000)),
    "RIVER": ((8000, 9500),  (8000, 9500)),
    "JUNGLE":((3000, 12000), (3000, 12000)),
}

def in_box(x, y, box):
    (xmin, xmax), (ymin, ymax) = box
    return xmin <= x <= xmax and ymin <= y <= ymax

def lane_from_position(x, y):
    if in_box(x,y, LANE_BOXES["TOP"]):   return "TOP_area"
    if in_box(x,y, LANE_BOXES["BOT"]):   return "BOT_area"
    if in_box(x,y, LANE_BOXES["MID"]):   return "MID_area"
    if in_box(x,y, LANE_BOXES["RIVER"]): return "RIVER_area"
    if in_box(x,y, LANE_BOXES["JUNGLE"]):return "JUNGLE_area"
    return "OTHER"

def extract_item_build_from_timeline(timeline, participant_id):
    """ Devuelve (primer_mitico, core_items[:3], tiempo_mitico_seg) usando eventos ITEM_PURCHASED. """
    if not timeline: return (None, [], None)
    mythic = None
    mythic_time = None
    purchases = []
    for frame in timeline.get("info", {}).get("frames", []):
        ts = frame["timestamp"] / 1000.0
        for ev in frame.get("events", []):
            if ev.get("type") == "ITEM_PURCHASED" and ev.get("participantId") == participant_id:
                iid = ev.get("itemId")
                purchases.append((ts, iid))
                if iid in mythic_set and mythic is None:
                    mythic = iid
                    mythic_time = ts
    core = []
    skip_ids = {1001, 2055, 3340, 3364, 3363, 2138, 2139, 2140}
    for ts, iid in purchases:
        if iid in skip_ids:
            continue
        core.append(iid)
        core = list(dict.fromkeys(core))
        if len(core) >= 3: break
    return (mythic, core, mythic_time)

def summarize_player_from_match(match, timeline, puuid):
    info = match["info"]
    queue_id = info.get("queueId")
    mode = QUEUE_NAMES.get(queue_id, f"Queue {queue_id}")
    duration = info.get("gameDuration", 0)
    minutes = max(duration/60, 1)

    me = None
    team_kills = {100:0, 200:0}
    team_gold  = {100:0, 200:0}
    for p in info["participants"]:
        t = 100 if p["teamId"] == 100 else 200
        team_kills[t] += p.get("kills",0)
        team_gold[t]  += p.get("goldEarned",0)
        if p["puuid"] == puuid:
            me = p
    if me is None:
        return None

    t_id = me["teamId"]
    part_id = me["participantId"]

    mythic_iid, core_iids, mythic_time = extract_item_build_from_timeline(timeline, part_id)
    mythic_name = id2name.get(mythic_iid, None) if mythic_iid else None
    core_names  = [id2name.get(i, str(i)) for i in core_iids]

    kills, deaths, assists = me["kills"], me["deaths"], me["assists"]
    kda   = (kills + assists) / (deaths if deaths>0 else 1)

    cs      = me.get("totalMinionsKilled",0) + me.get("neutralMinionsKilled",0)
    csmin   = cs / minutes
    dchamps = me.get("totalDamageDealtToChampions",0)
    dpm     = dchamps / minutes
    gpm     = me.get("goldEarned",0) / minutes
    vscore  = me.get("visionScore",0)
    vpm     = vscore / minutes

    kp = (kills + assists) / max(team_kills[t_id], 1)
    dmg_share  = me.get("challenges",{}).get("teamDamagePercentage", np.nan)
    gold_share = me.get("goldEarned",0) / max(team_gold[t_id], 1)

    ch = me.get("challenges", {})
    obj_part = {
        "dragons": ch.get("dragonTakedowns", 0),
        "barons": ch.get("baronTakedowns", 0),
        "heralds": ch.get("riftHeraldTakedowns", 0),
        "towers": me.get("turretKills", 0),
        "inhibitors": me.get("inhibitorKills", 0),
        "soloKills": ch.get("soloKills", 0),
        "wardsPlaced": me.get("wardsPlaced", 0),
        "wardsKilled": me.get("wardsKilled", 0),
        "controlWards": me.get("visionWardsBoughtInGame", 0),
        "skillshotsHit": ch.get("skillshotsHit", np.nan),
        "skillshotsDodged": ch.get("skillshotsDodged", np.nan),
    }

    role = me.get("teamPosition", me.get("role","UNKNOWN"))

    roam_pct = np.nan
    dist_traveled = np.nan
    lane_time = {}
    if timeline:
        xs, ys = [], []
        lane_bins = {"TOP_area":0, "MID_area":0, "BOT_area":0, "RIVER_area":0, "JUNGLE_area":0, "OTHER":0}
        last_x, last_y = None, None
        dist = 0.0
        frames = timeline.get("info", {}).get("frames", [])
        for f in frames:
            ps = f.get("participantFrames", {}).get(str(part_id), {})
            if "position" in ps:
                x = ps["position"]["x"]; y = ps["position"]["y"]
                xs.append(x); ys.append(y)
                if last_x is not None:
                    dist += math.dist((x,y), (last_x,last_y))
                last_x, last_y = x, y
                lane_bins[lane_from_position(x,y)] += 1
        tot_frames = sum(lane_bins.values())
        if tot_frames>0:
            lane_time = {k: v/tot_frames for k,v in lane_bins.items()}
            lane_map = {
                "TOP":"TOP_area","MIDDLE":"MID_area","MID":"MID_area",
                "BOTTOM":"BOT_area","ADC":"BOT_area","CARRY":"BOT_area",
                "UTILITY":"BOT_area","JUNGLE":"JUNGLE_area"
            }
            main_area = lane_map.get(role, None)
            if main_area:
                roam_pct = 1 - lane_time.get(main_area, 0)
            dist_traveled = dist

    win = bool(me.get("win", False))

    return {
        "matchId": match["metadata"]["matchId"],
        "queueId": queue_id,
        "mode": mode,
        "win": win,
        "champion": me["championName"],
        "role": role,
        "gameMinutes": round(minutes,2),
        "kills": kills,
        "deaths": deaths,
        "assists": assists,
        "KDA": round(kda,2),
        "CS": cs,
        "CS/min": round(csmin,2),
        "DmgChamp": dchamps,
        "DPM": round(dpm,2),
        "Gold": me.get("goldEarned",0),
        "GPM": round(gpm,2),
        "Vision": vscore,
        "Vision/min": round(vpm,2),
        "KP": round(kp,2),
        "%TeamDamage": round(dmg_share,2) if pd.notna(dmg_share) else np.nan,
        "%TeamGold": round(gold_share,2),
        "soloKills": obj_part["soloKills"],
        "wardsPlaced": obj_part["wardsPlaced"],
        "wardsKilled": obj_part["wardsKilled"],
        "controlWards": obj_part["controlWards"],
        "dragons": obj_part["dragons"],
        "barons": obj_part["barons"],
        "heralds": obj_part["heralds"],
        "towers": obj_part["towers"],
        "inhibitors": obj_part["inhibitors"],
        "skillshotsHit": obj_part["skillshotsHit"],
        "skillshotsDodged": obj_part["skillshotsDodged"],
        "mythic": mythic_name,
        "coreItems": ", ".join(core_names) if core_names else "",
        "mythicTime_s": round(mythic_time,1) if mythic_time else np.nan,
        "roamPct": round(roam_pct,2) if pd.notna(roam_pct) else np.nan,
        "distanceTraveled": round(dist_traveled,1) if pd.notna(dist_traveled) else np.nan
    }

# ===========================
# 8) DESCARGA MASIVA Y TABLA PRINCIPAL
# ===========================
def build_dataset(match_ids, puuid, with_timeline=True, desc="Descargando"):
    rows = []
    for mid in tqdm(match_ids, desc=desc):
        try:
            m,t = fetch_match(mid, with_timeline=with_timeline)
            row = summarize_player_from_match(m,t,puuid)
            if row:
                rows.append(row)
        except Exception as e:
            # print("Error en", mid, e)
            pass
    return pd.DataFrame(rows)

df_ranked = build_dataset(match_ids_ranked, puuid, with_timeline=True, desc="Ranked (con timeline)")
df_normal = build_dataset(match_ids_normal, puuid, with_timeline=True, desc="Normales (con timeline)")

df_flex = df_ranked[df_ranked["queueId"] == 440].copy()
df_solo = df_ranked[df_ranked["queueId"] == 420].copy()

print("Partidas Flex:", len(df_flex), "| Solo/Duo:", len(df_solo), "| Normales:", len(df_normal))

df_all = pd.concat([
    df_flex.assign(modeGroup="Flex"),
    df_solo.assign(modeGroup="SoloQ"),
    df_normal.assign(modeGroup="Normal")
], ignore_index=True)

display(df_all.head(3))

# ===========================
# 9) RESÚMENES
# ===========================
def basic_summary(df, title):
    if df.empty:
        print(f"[{title}] Sin partidas.")
        return None
    g = {
        "N": len(df),
        "Winrate": round(df["win"].mean()*100, 1),
        "KDA": round((df["kills"]+df["assists"]).sum() / max(df["deaths"].sum(), 1), 2),
        "CS/min": round(df["CS"].sum() / max(df["gameMinutes"].sum(),1), 2),
        "DPM": round(df["DmgChamp"].sum() / max(df["gameMinutes"].sum(),1), 1),
        "GPM": round(df["Gold"].sum() / max(df["gameMinutes"].sum(),1), 1),
        "Vision/min": round(df["Vision"].sum() / max(df["gameMinutes"].sum(),1), 2),
        "KP prom": round(df["KP"].mean()*100,1),
        "%TeamDamage prom": round(df["%TeamDamage"].mean()*100,1),
        "SoloKills prom": round(df["soloKills"].mean(),2),
        "Roam% prom": round(df["roamPct"].mean()*100,1) if "roamPct" in df else np.nan,
        "Mythic más usado": df["mythic"].mode().iloc[0] if df["mythic"].dropna().size>0 else None
    }
    print(f"\n=== {title} ===")
    return pd.DataFrame([g])

sum_flex  = basic_summary(df_flex,  "Resumen FLEX")
sum_solo  = basic_summary(df_solo,  "Resumen SOLO/DUO")
sum_norm  = basic_summary(df_normal,"Resumen NORMALES")
display(pd.concat([s for s in [sum_flex,sum_solo,sum_norm] if s is not None], ignore_index=True))

# ===========================
# 10) MEJORES CAMPEONES / ROLES
# ===========================
def top_champions(df, k=10):
    if df.empty: return pd.DataFrame()
    agg = (df.groupby("champion")
             .agg(N=("win","size"),
                  Winrate=("win","mean"),
                  CSmin=("CS/min","mean"),
                  DPM=("DPM","mean"),
                  KDA=("KDA","mean"),
                  KP=("KP","mean"),
                  Vision_min=("Vision/min","mean"))
             .reset_index())
    agg["Winrate"] = (agg["Winrate"]*100).round(1)
    return agg.sort_values(["Winrate","N"], ascending=[False,False]).head(k)

print("\nTop campeones FLEX")
display(top_champions(df_flex, 10))
print("\nTop campeones NORMALES")
display(top_champions(df_normal, 10))

def role_performance(df):
    if df.empty: return pd.DataFrame()
    agg = (df.groupby("role")
             .agg(N=("win","size"),
                  Winrate=("win","mean"),
                  KDA=("KDA","mean"),
                  CSmin=("CS/min","mean"),
                  DPM=("DPM","mean"),
                  Vision_min=("Vision/min","mean"),
                  Roam_pct=("roamPct","mean"))
             .reset_index())
    agg["Winrate"] = (agg["Winrate"]*100).round(1)
    agg["Roam_pct"] = (agg["Roam_pct"]*100).round(1)
    return agg.sort_values(["Winrate","N"], ascending=[False,False])

print("\nRendimiento por ROL en FLEX")
display(role_performance(df_flex))
print("\nRendimiento por ROL en NORMALES")
display(role_performance(df_normal))

# ===========================
# 11) BUILDS (Mítico + Núcleo)
# ===========================
def build_stats(df, k=10):
    if df.empty: return pd.DataFrame()
    tmp = df.copy()
    tmp["coreKey"] = tmp["coreItems"].fillna("").replace("", "None")
    agg = (tmp.groupby(["champion","mythic","coreKey"])
             .agg(N=("win","size"), Winrate=("win","mean"), KDA=("KDA","mean"),
                  DPM=("DPM","mean"), CSmin=("CS/min","mean"))
             .reset_index())
    agg["Winrate"] = (agg["Winrate"]*100).round(1)
    return agg.sort_values(["Winrate","N"], ascending=[False,False]).head(k)

print("\nMejores builds en FLEX (por WR y N)")
display(build_stats(df_flex, 15))
print("\nMejores builds en NORMALES (por WR y N)")
display(build_stats(df_normal, 15))

# ===========================
# 12) HEATMAP DE MOVIMIENTO
# ===========================
def movement_heatmap(df, title="Heatmap de movimiento", sample=10):
    mids = df["matchId"].drop_duplicates().head(sample).tolist()
    xs_all, ys_all = [], []
    for mid in mids:
        try:
            timeline = safe_get(f"{BASE_M5}/matches/{mid}/timeline")
            m = safe_get(f"{BASE_M5}/matches/{mid}")
            me = next((p for p in m["info"]["participants"] if p["puuid"] == puuid), None)
            if not me:
                continue
            pid = me["participantId"]
            for fr in timeline["info"]["frames"]:
                pf = fr.get("participantFrames",{}).get(str(pid),{})
                if "position" in pf:
                    xs_all.append(pf["position"]["x"])
                    ys_all.append(pf["position"]["y"])
        except Exception:
            pass
    if not xs_all:
        print("Sin datos de posición para heatmap.")
        return
    plt.figure(figsize=(6,6))
    hb = plt.hexbin(xs_all, ys_all, gridsize=50, bins='log')
    plt.title(title)
    plt.xlabel("X map"); plt.ylabel("Y map")
    plt.colorbar(hb, label="densidad (log)")
    plt.show()

print("\nHeatmap FLEX (muestra 8 partidas)")
movement_heatmap(df_flex, "Movimiento - FLEX", sample=8)
print("\nHeatmap NORMALES (muestra 8 partidas)")
movement_heatmap(df_normal, "Movimiento - NORMALES", sample=8)

# ===========================
# 13) GRÁFICOS CLAVE
# ===========================
def plot_metric(df, metric, title, top_k=15):
    if df.empty:
        print("Sin partidas.")
        return
    order = df.groupby("champion")[metric].mean().sort_values(ascending=False).head(top_k).index
    plt.figure(figsize=(8,4))
    sns.barplot(data=df[df["champion"].isin(order)], x="champion", y=metric, order=order)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

plot_metric(df_flex, "DPM", "Top DPM por campeón (FLEX)")
plot_metric(df_flex, "CS/min", "Top CS/min por campeón (FLEX)")
plot_metric(df_flex, "Vision/min", "Top Visión/min por campeón (FLEX)")

plot_metric(df_normal, "DPM", "Top DPM por campeón (NORMALES)")
plot_metric(df_normal, "CS/min", "Top CS/min por campeón (NORMALES)")
plot_metric(df_normal, "Vision/min", "Top Visión/min por campeón (NORMALES)")

# ===========================
# 14) SUGERENCIAS DE MEJORA
# ===========================
def improvement_hints(df, mode_label=""):
    if df.empty:
        return pd.DataFrame()
    tips = []
    if df["KDA"].mean() < 2:
        tips.append("Posicionamiento: KDA bajo (<2). Juega más seguro en mid game y usa pings defensivos.")
    cs_min = (df["CS"].sum() / df["gameMinutes"].sum())
    if cs_min < 6:
        tips.append("CS/min bajo (<6). Practica last hit; apunta a 7–8 cs/min en línea.")
    vis_min = (df["Vision"].sum() / df["gameMinutes"].sum())
    if vis_min < 0.6:
        tips.append("Poca visión (<0.6/min). Compra más pinks y renueva wards en objetivos.")
    if df["KP"].mean() < 0.45:
        tips.append("Baja participación en kills (<45%). Coordina rotaciones y juega alrededor de tu jungla.")
    if df["dragons"].mean() < 0.3 and df["barons"].mean() == 0:
        tips.append("Poca participación en objetivos. Llega con tiempo a visión previa (1:00 antes).")
    if df["mythic"].dropna().size>0 and df["win"].mean()<0.5:
        tips.append("Revisa tu mítico/primeros 3 items: prueba la variante con mayor WR en tus campeones top.")
    if "roamPct" in df and df["roamPct"].mean() > 0.45 and df["CS/min"].mean()<6:
        tips.append("Roaming excesivo con bajo CS. Prioriza empujar oleada antes de moverte.")
    return pd.DataFrame({"Sugerencias "+mode_label: tips})

print("\nSugerencias para FLEX")
display(improvement_hints(df_flex, "(Flex)"))
print("\nSugerencias para NORMALES")
display(improvement_hints(df_normal, "(Normales)"))

# ===========================
# 15) INFORME FINAL (ORDEN LÓGICO)
# ===========================
report_modes = pd.concat(
    [s.assign(MODE=lab) for s,lab in [(sum_flex,"Flex"), (sum_norm,"Normal")] if s is not None],
    ignore_index=True
) if (sum_flex is not None or sum_norm is not None) else None

print("\n=== 1) Panorama por modo ===")
if report_modes is not None:
    display(report_modes)

print("\n=== 2) Mejores campeones por desempeño ===")
print("FLEX:");   display(top_champions(df_flex, 10))
print("NORMALES:");display(top_champions(df_normal, 10))

print("\n=== 3) Rendimiento por rol ===")
print("FLEX:");   display(role_performance(df_flex))
print("NORMALES:");display(role_performance(df_normal))

print("\n=== 4) Builds (Mítico + Núcleo) con mejor WR ===")
print("FLEX:");   display(build_stats(df_flex, 15))
print("NORMALES:");display(build_stats(df_normal, 15))

print("\n=== 5) Hábitos de juego (visión, KP, roaming, objetivos) ===")
cols_habitos = ["mode","champion","role","win","KDA","CS/min","DPM","Vision/min","KP","%TeamDamage","soloKills","dragons","barons","heralds","roamPct","mythic","coreItems"]
display(df_all[cols_habitos].sort_values("mode").head(25))

print("\n=== 6) Visualizaciones clave generadas arriba (barras + heatmaps) ===")
print("\nFIN DEL NOTEBOOK ✅")
```

```python
# ===========================
# 16) GRÁFICAS ADICIONALES
# ===========================
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="notebook", style="whitegrid")

def _nonempty(df):
    return df is not None and not df.empty

# ---------- 16.1 Distribuciones (hist/kde) ----------
def plot_distributions(df, title):
    if not _nonempty(df):
        print(f"[{title}] Sin partidas.")
        return
    metrics = ["KDA","CS/min","DPM","Vision/min","KP","%TeamDamage","roamPct"]
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(14,10))
    axes = axes.flatten()
    for i, m in enumerate(metrics):
        if m in df.columns:
            sns.histplot(df[m].dropna(), kde=True, ax=axes[i])
            axes[i].set_title(m)
    fig.suptitle(f"Distribuciones - {title}", fontsize=14)
    plt.tight_layout()
    plt.show()

plot_distributions(df_flex, "FLEX")
plot_distributions(df_normal, "NORMALES")

# ---------- 16.2 Boxplots por rol ----------
def boxplots_by_role(df, title):
    if not _nonempty(df): return
    metrics = ["KDA","CS/min","DPM","Vision/min","KP","roamPct"]
    for m in metrics:
        if m in df.columns:
            plt.figure(figsize=(7,4))
            sns.boxplot(data=df, x="role", y=m, order=sorted(df["role"].dropna().unique()))
            plt.title(f"{m} por rol - {title}")
            plt.xlabel("Rol"); plt.ylabel(m)
            plt.tight_layout()
            plt.show()

boxplots_by_role(df_flex, "FLEX")
boxplots_by_role(df_normal, "NORMALES")

# ---------- 16.3 Scatter: relaciones clave ----------
def scatter_relations(df, title):
    if not _nonempty(df): return
    # KDA vs CS/min (tamaño = KP, color = rol)
    if {"KDA","CS/min","KP","role"}.issubset(df.columns):
        plt.figure(figsize=(7,5))
        sns.scatterplot(data=df, x="CS/min", y="KDA", hue="role",
                        size="KP", sizes=(20,200), alpha=0.8)
        plt.title(f"KDA vs CS/min — {title} (tamaño = KP)")
        plt.tight_layout(); plt.show()
    # DPM vs GPM
    if {"DPM","GPM","role"}.issubset(df.columns):
        plt.figure(figsize=(7,5))
        sns.scatterplot(data=df, x="GPM", y="DPM", hue="role", alpha=0.9)
        plt.title(f"DPM vs GPM — {title}")
        plt.tight_layout(); plt.show()
    # KP vs Vision/min
    if {"KP","Vision/min","role"}.issubset(df.columns):
        plt.figure(figsize=(7,5))
        sns.scatterplot(data=df, x="Vision/min", y="KP", hue="role", alpha=0.9)
        plt.title(f"KP vs Visión/min — {title}")
        plt.tight_layout(); plt.show()

scatter_relations(df_flex, "FLEX")
scatter_relations(df_normal, "NORMALES")

# ---------- 16.4 Tendencias por partida (orden cronológico aproximado por índice) ----------
def trends_by_match(df, title):
    if not _nonempty(df): return
    df_idx = df.sort_values("matchId").reset_index(drop=True).copy()
    df_idx["match#"] = np.arange(1, len(df_idx)+1)
    for m in ["KDA","CS/min","DPM","Vision/min","KP"]:
        if m in df_idx.columns:
            plt.figure(figsize=(8,3.5))
            sns.lineplot(data=df_idx, x="match#", y=m, marker="o")
            plt.title(f"Tendencia de {m} por partida — {title}")
            plt.xlabel("Partida (orden aproximado)"); plt.ylabel(m)
            plt.tight_layout(); plt.show()

trends_by_match(df_flex, "FLEX")
trends_by_match(df_normal, "NORMALES")

# ---------- 16.5 Campeones top: Winrate y KDA ----------
def bar_top_champs(df, title, top_k=10):
    if not _nonempty(df): return
    agg = (df.groupby("champion")
             .agg(N=("win","size"), WR=("win","mean"), KDA=("KDA","mean"))
             .reset_index())
    agg["WR"] = (agg["WR"]*100).round(1)
    agg = agg.sort_values(["WR","N"], ascending=[False, False]).head(top_k)
    fig, axes = plt.subplots(1,2, figsize=(12,4))
    sns.barplot(data=agg, x="champion", y="WR", ax=axes[0])
    axes[0].set_title(f"Top WR — {title}"); axes[0].set_xlabel(""); axes[0].set_ylabel("Winrate (%)")
    axes[0].tick_params(axis='x', rotation=45)
    sns.barplot(data=agg, x="champion", y="KDA", ax=axes[1])
    axes[1].set_title(f"KDA medio — {title}"); axes[1].set_xlabel(""); axes[1].set_ylabel("KDA")
    axes[1].tick_params(axis='x', rotation=45)
    plt.tight_layout(); plt.show()

bar_top_champs(df_flex, "FLEX", 10)
bar_top_champs(df_normal, "NORMALES", 10)

# ---------- 16.6 Violin: Roaming por rol ----------
def violin_roam(df, title):
    if not _nonempty(df) or "roamPct" not in df.columns: return
    plt.figure(figsize=(7,4))
    sns.violinplot(data=df, x="role", y="roamPct", inner="quartile",
                   order=sorted(df["role"].dropna().unique()))
    plt.title(f"Distribución de Roam% por rol — {title}")
    plt.xlabel("Rol"); plt.ylabel("Roam%")
    plt.tight_layout(); plt.show()

violin_roam(df_flex, "FLEX")
violin_roam(df_normal, "NORMALES")

# ---------- 16.7 Uso de míticos y rendimiento ----------
def mythic_usage(df, title, top_k=12):
    if not _nonempty(df) or "mythic" not in df.columns: return
    tmp = df.copy()
    tmp["mythic"] = tmp["mythic"].fillna("None")
    agg = (tmp.groupby("mythic")
           .agg(N=("win","size"), WR=("win","mean"), KDA=("KDA","mean"), DPM=("DPM","mean"))
           .reset_index())
    agg["WR"] = (agg["WR"]*100).round(1)
    agg = agg.sort_values("N", ascending=False).head(top_k)
    fig, axes = plt.subplots(1,2, figsize=(12,4))
    sns.barplot(data=agg, x="mythic", y="N", ax=axes[0])
    axes[0].set_title(f"Uso de míticos — {title}"); axes[0].tick_params(axis='x', rotation=45)
    sns.barplot(data=agg, x="mythic", y="WR", ax=axes[1])
    axes[1].set_title(f"Winrate por mítico — {title}"); axes[1].tick_params(axis='x', rotation=45)
    axes[1].set_ylabel("Winrate (%)")
    plt.tight_layout(); plt.show()

mythic_usage(df_flex, "FLEX")
mythic_usage(df_normal, "NORMALES")

# ---------- 16.8 Objetivos por partida (promedio) ----------
def objectives_bars(df, title):
    if not _nonempty(df): return
    cols = ["dragons","barons","heralds","towers","inhibitors"]
    cols = [c for c in cols if c in df.columns]
    means = df[cols].mean().sort_values(ascending=False)
    plt.figure(figsize=(7,4))
    sns.barplot(x=means.index, y=means.values)
    plt.title(f"Participación en objetivos — {title}")
    plt.ylabel("Promedio por partida"); plt.xlabel("")
    plt.tight_layout(); plt.show()

objectives_bars(df_flex, "FLEX")
objectives_bars(df_normal, "NORMALES")

# ---------- 16.9 Radar por campeón (perfil de juego) ----------
def radar_for_champion(df, champion, title_suffix=""):
    if not _nonempty(df): return
    subset = df[df["champion"] == champion]
    if subset.empty:
        print(f"{champion}: sin datos.")
        return
    # métricas normalizadas vs promedio global
    metrics = ["KDA","CS/min","DPM","Vision/min","KP","%TeamDamage","roamPct"]
    subset_mean = subset[metrics].mean(numeric_only=True)
    global_mean = df[metrics].mean(numeric_only=True)
    ratio = (subset_mean / global_mean).fillna(1.0)  # >1 mejor que tu media global
    # Radar
    labels = metrics
    values = ratio.values.tolist()
    values += values[:1]
    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
    angles += angles[:1]
    fig = plt.figure(figsize=(6,6))
    ax = plt.subplot(111, polar=True)
    ax.plot(angles, values, linewidth=2)
    ax.fill(angles, values, alpha=0.25)
    ax.set_thetagrids(np.degrees(angles[:-1]), labels)
    ax.set_title(f"Radar: {champion} (>{title_suffix})\n>1 = por encima de tu media")
    ax.set_rlim(0, max(1.6, np.nanmax(values)+0.1))
    plt.show()

# Radar de tus 3 campeones más jugados en cada modo
def radar_top3(df, mode_name):
    if not _nonempty(df): return
    top3 = df["champion"].value_counts().head(3).index.tolist()
    for ch in top3:
        radar_for_champion(df, ch, title_suffix=mode_name)

radar_top3(df_flex, "FLEX")
radar_top3(df_normal, "NORMALES")

# ---------- 16.10 Pairplot rápido (cuidado: pesado si hay muchas partidas) ----------
def quick_pairplot(df, title, sample=120):
    if not _nonempty(df): return
    use = df.sample(min(len(df), sample), random_state=42)
    cols = ["KDA","CS/min","DPM","Vision/min","KP","%TeamDamage"]
    cols = [c for c in cols if c in use.columns]
    if len(cols) < 2:
        return
    g = sns.pairplot(use[cols], corner=True)
    g.fig.suptitle(f"Pairplot (muestra n={len(use)}) — {title}", y=1.02)
    plt.show()

quick_pairplot(df_flex, "FLEX")
quick_pairplot(df_normal, "NORMALES")

print("✅ Gráficas adicionales generadas.")
```
